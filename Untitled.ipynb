## Feature Understanding

Also Known as univariate analysis

```python
df['year_introduced'].value_counts()
```

#### Let's Make a cluster of all the years, made less than 20 roller coasters

```python
coasters = df['year_introduced'].value_counts()
coaster_lessThan20 = coasters[coasters <= 20]
```

```python
def handle_year(value):
    if value in coaster_lessThan20:
        return 'years made less than 20 coasters'
    else:
        return value

df['year_introduced'] = df['year_introduced'].apply(handle_year)                          
```

```python
plt.figure(figsize = (10, 6))
sns.countplot(x = 'year_introduced', data = df)
plt.xticks(rotation = 90)
```
